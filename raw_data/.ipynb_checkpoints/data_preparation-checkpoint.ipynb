## Generic Settings and Imports

```python
from pathlib import Path

import GPy
import matplotlib
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from prophet import Prophet
```

```python
%matplotlib inline

matplotlib.rcParams.update({
    # for nice exports to LaTeX
    "pgf.texsystem": "pdflatex",
    'font.family': 'serif',
    'text.usetex': True,
    'pgf.rcfonts': False,
    # standard figure size nice for A4 pages.
    # 'figure.figsize': [6.3, 3.5],
    # standard figure size for double columns
    'figure.figsize': [3.3, 1.8],
    'font.size': 9,
    # Makes plotted stuff larger in notebook
    'figure.dpi': 100,
})
```

## Generic Helpers

```python
def add_minute_of_day(df):
    df["minutes of day"] = df.index.hour * 60 + df.index.minute
    return df
```

```python
def save_ax_for_latex(ax, fn):
    """
    Allows saving the stuff returned by a pandas plot.
    """
    fig = ax.get_figure()
    plt.tight_layout()
    
    base_dir = Path().absolute()

    # Save somewhere else on David's machine.
    if "diss" in str(base_dir):
        base_dir = Path(str(base_dir).split("diss")[0])
        base_dir = base_dir / "diss" / "generated_graphics"
        
    fig.savefig(str(base_dir / fn), bbox_inches='tight')
```

## Load the Raw Data

```python
data_url = "https://raw.githubusercontent.com/fzi-forschungszentrum-informatik/tropical_precooling_environment/master/tropical_precooling/data/building_measurements.csv"
raw_data = pd.read_csv(data_url, parse_dates=True, index_col=0)
raw_data
```

## Prepare the Data for the Predictions
In particular add the minutes of day as a feature for the forecast of the supply air temperature 

```python
data = add_minute_of_day(raw_data.copy())
data
```

```python
_ = pd.plotting.scatter_matrix(data, figsize=(10,10))
```

## Train-Test Split as in the Tropcial Precooling Env

```python
train_selector = data.index.month >= 7
predict_selector = data.index.month > 0
metrics_selector = data.index.month < 7
```

## Predict the supply air temperature

```python
# all other data is not available in advance.
sat_features = data[["minutes of day", "Outside air temperature"]]

# normalize data
sat_features = (sat_features - sat_features.mean()) / sat_features.std()

X_train = sat_features[train_selector].values
print(X_train.shape, X_train.mean(), X_train.std())
X_predict = sat_features[predict_selector].values
print(X_predict.shape, X_predict.mean(), X_predict.std())

Y_train = data[["Supply air temp"]].loc[train_selector].values
print(Y_train.shape, Y_train.mean(), Y_train.std())
Y_true = data[["Supply air temp"]].loc[predict_selector].values
print(Y_true.shape, Y_true.mean(), Y_true.std())
```

```python
kernel = GPy.kern.RBF(input_dim=2, variance=1., lengthscale=1., ARD=True)
mean_function = GPy.mappings.Constant(input_dim=2, output_dim=1, value=Y_train.mean())
gp = GPy.models.GPRegression(X_train, Y_train, kernel, mean_function=mean_function)
```

```python
_ = gp.optimize(messages=True)
```

```python
Y_predict, Y_predict_var = gp.predict(X_predict)
```

```python
Y_predict_std = np.sqrt(Y_predict_var)
sat_prediction = pd.DataFrame(
    index=data.index[predict_selector],
    data={
        "sat_fct_mean": Y_predict[:, 0],
        "sat_fct_lower": Y_predict[:, 0] - 2 * Y_predict_std[:, 0],
        "sat_fct_upper": Y_predict[:, 0] + 2 * Y_predict_std[:, 0],
    },
)
sat_prediction
```

```python
sat_metrics = pd.DataFrame(
    index=data.index,
    data={
        "MAE": abs(data["Supply air temp"] - sat_prediction["sat_fct_mean"]),
        "MAPE": (abs(data["Supply air temp"] - sat_prediction["sat_fct_mean"])/sat_prediction["sat_fct_mean"]),
    },
)
sat_metrics = sat_metrics[metrics_selector]
ax = sat_metrics.plot(secondary_y=["MAPE"])
```

```python
sat_metrics
```

```python
sat_metrics.groupby(lambda x: x.time()).mean().plot(secondary_y=["MAPE"])
```

```python
sat_metrics.mean()
```

## Forecast for $CO_2$ values

```python
train_data = pd.DataFrame(
    data={
        "ds": data[train_selector].index,
        "y": data[train_selector]["CO2"]
    }
)
predict_data = pd.DataFrame(
    data={
        "ds": data[predict_selector].index,
    }
)

co2_model = Prophet(growth="flat", n_changepoints=0, yearly_seasonality=False, interval_width=0.954)
_ = co2_model.fit(train_data)
```

```python
prophet_out = co2_model.predict(predict_data)
```

```python
co2_model.plot(prophet_out)
```

```python
co2_model.plot_components(prophet_out)
```

```python
prophet_out
```

```python
co2_prediction = pd.DataFrame(
    index=prophet_out["ds"].values,
    data={
        "co2_fct_mean": prophet_out["yhat"].values,
        "co2_fct_upper": prophet_out["yhat_upper"].values,
        "co2_fct_lower": prophet_out["yhat_lower"].values,
    },
)
co2_prediction
```

```python
co2_metrics = pd.DataFrame(
    index=data.index[predict_selector],
    data={
        "MAE": abs(data["CO2"] - co2_prediction["co2_fct_mean"]),
        "MAPE": (abs(data["CO2"] - co2_prediction["co2_fct_mean"])/co2_prediction["co2_fct_mean"]),
    },
)
co2_metrics = co2_metrics[metrics_selector]
ax = co2_metrics.plot(secondary_y=["MAPE"])
```

```python
ax = co2_metrics.groupby(lambda x: x.time()).mean().plot(secondary_y=["MAPE"])
save_ax_for_latex(ax, "dummy_fct_error.pdf")
```

```python
co2_metrics.mean()
```
