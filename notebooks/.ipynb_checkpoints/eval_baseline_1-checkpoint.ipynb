# Evaluation of Performance of Baseline 1

```python
from pathlib import Path

from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
from scipy.linalg import inv
from tqdm import tqdm
```

```python
from tbmenv import Scenario1
from tbmenv import Baseline1
```

## Performance with Default Parameters

```python
env = Scenario1()
model = Baseline1()

model.observe(*env.get_training_data())
model.observe(*env.reset())

done = False
pbar = tqdm(total=env.n_predictions_expected)
while not done:
    predicted_states = model.predict(use_tuned_variables=False)
    obs, actions, disturbances, done = env.step(predicted_states)
    model.observe(obs, actions, disturbances)
    pbar.update(1)
pbar.close()

residuals = env.compute_performance_measure(return_per_lead_step=True)
mae_per_step = residuals.mean(axis=0).mean(axis=-1)
mae_per_step = pd.DataFrame(
    data=mae_per_step,
    columns=["Default Parameters"]
)
mape = env.compute_mape()
mape_per_step = mape.mean(axis=0).mean(axis=-1)
mape_per_step = pd.DataFrame(
    data=mape_per_step,
    columns=["Default Parameters"]
)

print("MAE: {:.2f}°C".format(mae_per_step.values.mean()))
print("MAPE: {:.1f}%".format(mape_per_step.values.mean()))
mae_df = pd.DataFrame(data=mae_per_step)
mae_df.index.name = "Time Steps of Prediction"
mape_df = pd.DataFrame(data=mape_per_step)
mape_df.index.name = "Time Steps of Prediction"
_ = mae_df.plot(ylabel = "MAE [°C]")
_ = mape_df.plot(ylabel = "MAPE [%]")
```

## Performance with Tuned Parameters

```python
env = Scenario1()
model = Baseline1()

model.observe(*env.get_training_data())
model.observe(*env.reset())

done = False
pbar = tqdm(total=env.n_predictions_expected)
while not done:
    predicted_states = model.predict(use_tuned_variables=True)
    obs, actions, disturbances, done = env.step(predicted_states)
    model.observe(obs, actions, disturbances)
    pbar.update(1)
pbar.close()

residuals = env.compute_performance_measure(return_per_lead_step=True)
mae_per_step = residuals.mean(axis=0).mean(axis=-1)
mae_per_step = pd.DataFrame(
    data=mae_per_step,
    columns=["Tuned Parameters"]
)
mape = env.compute_mape()
mape_per_step = mape.mean(axis=0).mean(axis=-1)
mape_per_step = pd.DataFrame(
    data=mape_per_step,
    columns=["Tuned Parameters"]
)

print("MAE: {:.2f}°C".format(mae_per_step.values.mean()))
print("MAPE: {:.1f}%".format(mape_per_step.values.mean()))
mae_df = pd.concat([mae_df, mae_per_step])
mape_df = pd.concat([mape_df, mape_per_step])
_ = mae_df.plot(ylabel = "MAE [°C]")
_ = mape_df.plot(ylabel = "MAPE [%]")
```
