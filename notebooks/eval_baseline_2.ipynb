# Evaluation of Performance of Baseline 2

```python
from matplotlib import pyplot as plt
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
from tbmenv import Scenario2
from tbmenv import Baseline2
```

## Performance Check

```python
env = Scenario2()
model = Baseline2()

model.observe(*env.get_training_data())
model.observe(*env.reset())

done = False
while not done:
    predicted_states = model.predict(n_dates=1)
    obs, actions, disturbances, done = env.step(predicted_states)
    model.observe(obs, actions, disturbances)

env.compute_performance_measure()
```

```python
residuals = env.compute_performance_measure(return_per_lead_step=True)
mape = env.compute_mape()
mae_per_step = residuals.mean(axis=0)
mape_per_step = mape.mean(axis=0)

print("MAE: {:.2f}Â°C".format(mae_per_step.mean()))
print("MAPE: {:.1f}%".format(mape_per_step.mean()))

mae_df = pd.DataFrame(data=mae_per_step, columns=["w/ predicted disturbances"])
mae_df.index.name = "Time Steps of Prediction"
_ = mae_df.plot(ylabel = "MAE [Â°C]")
mape_df = pd.DataFrame(data=mape_per_step, columns=["w/ predicted disturbances"])
mape_df.index.name = "Time Steps of Prediction"
_ = mape_df.plot(ylabel = "MAPE [Â°C]")
```

### Finding:
The performance is acutally quite good. The spike at around the 25th step is likely due to the not observed setpoints of the AC system. I.e. the data show (see the data preparation notebook), that the AC is not switched on at the same time each day, but it slightly varies around half an hour or so. However, as the setpoint has not been recorded we cannot know with certainty.

## Check the influence of the number of historic dates used for prediction

```python
all_pm = []
all_n_dates = []
for n_dates in tqdm(range(1, 16)):
    # Reset to allow us computing from start without loosing the cache.
    model.obs_all = {}
    model.actions_all = {}
    model.disturbances_all = {}

    # Remaining code is as usual.
    model.observe(*env.get_training_data())
    model.observe(*env.reset())

    done = False
    while not done:
        predicted_states = model.predict(n_dates=n_dates)
        obs, actions, disturbances, done = env.step(predicted_states)
        model.observe(obs, actions, disturbances)

    all_pm.append(env.compute_performance_measure())
    all_n_dates.append(n_dates)
pm_over_n_dates = pd.DataFrame(
    index=all_n_dates,
    data={"MAE w/ predicted_disturbances": all_pm},
)
_ = pm_over_n_dates.plot.bar()
```
